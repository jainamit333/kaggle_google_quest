```python
import pandas as pd
import pandas_profiling as pp
import numpy as np
from urllib.parse import urlparse
```

```python
data_path = '/Users/amitjain/personal/machine-learning/Machine_Learning/competitions/google_quest/data/google-quest-challenge/train.csv' 
```

```python
data = pd.read_csv(data_path)
```

```python
data.columns
```

```python
domain_features = ['question_user_page', 'answer_user_page', 'url', 'host']
```

```python
data[domain_features].head()
```

```python
data['question_user_page_domain'] = data.apply (lambda row: urlparse(row.question_user_page).netloc, axis=1)
data['answer_user_page_domain'] = data.apply (lambda row: urlparse(row.answer_user_page).netloc, axis=1)
data['url_domain'] = data.apply (lambda row: urlparse(row.url).netloc, axis=1)
```

```python
computed_features = ['question_user_page_domain', 'answer_user_page_domain', 'url_domain']
```

```python
data[np.concatenate([domain_features, computed_features])].head()
```

```python
print('1: ', data['question_user_page_domain'].equals(data['answer_user_page_domain']))
print('2: ', data['question_user_page_domain'].equals(data['url_domain']))
print('3: ', data['answer_user_page_domain'].equals(data['url_domain']))

print('4: ', data['question_user_page_domain'].equals(data['host']))
print('5: ', data['answer_user_page_domain'].equals(data['host']))
print('6: ', data['url_domain'].equals(data['host']))
```

```python
labels = ['question_asker_intent_understanding',
       'question_body_critical', 'question_conversational',
       'question_expect_short_answer', 'question_fact_seeking',
       'question_has_commonly_accepted_answer',
       'question_interestingness_others', 'question_interestingness_self',
       'question_multi_intent', 'question_not_really_a_question',
       'question_opinion_seeking', 'question_type_choice',
       'question_type_compare', 'question_type_consequence',
       'question_type_definition', 'question_type_entity',
       'question_type_instructions', 'question_type_procedure',
       'question_type_reason_explanation', 'question_type_spelling',
       'question_well_written', 'answer_helpful',
       'answer_level_of_information', 'answer_plausible', 'answer_relevance',
       'answer_satisfaction', 'answer_type_instructions',
       'answer_type_procedure', 'answer_type_reason_explanation',
       'answer_well_written']
```

```python
data['host'] = data['host'].astype('category')
```

```python
data["host_cat"] = data["host"].cat.codes
```

```python
pp.ProfileReport(data[np.concatenate([labels, ['host_cat']])])
```
